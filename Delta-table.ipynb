```
%sql
CREATE TEMPORARY VIEW covid_data
USING CSV
OPTIONS (path "/FileStore/shared_uploads/boutranacir@gmail.com/covid_data.csv", header "true", mode"FAILFAST")
```

```
%sql
CREATE OR REPLACE TABLE covid.covid_data_delta
USING DELTA
LOCATION "/FileStore/shared_uploads/boutranacir@gmail.com/covid_data_delta"
AS
SELECT iso_code,location,continent,date,new_deaths_per_million,people_fully_vaccinated,population 
FROM covid_data
```

```
%sql
DELETE FROM covid.covid_data_delta where population is
null or people_fully_vaccinated is null or new_deaths_per_million is null or location is null
```
